# Data Analysis Intership-Task 3

**Import Python Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Dataset**

```python
data=pd.read_csv("/content/Zomato data  (1).csv")
```

**Data Cleaning**

```python
data.info()
```

```python
data.dropna(axis=1)
data.dropna(axis=0)
```

```python
data.isnull()
```

 **Convert the Data Type of Column_Rate**

```python
def handleRate(value):
  value=str(value).split("/")
  value=value[0]
  return float(value)
data["rate"]=data["rate"].apply(handleRate)
data.head()
```

```python
data.head()
```

**Data Visualisation Using Python**

**Count Plot**

```python
sns.countplot(x=data["listed_in(type)"], palette=['red','green','blue','orange'])
plt.title("Resturant Type Graph")
plt.xlabel("Resturant Type")
plt.ylabel("Count")
plt.show()
```

## **Conculsion-Majority of Customers Falls In Dining Category**

**Line PLot**

```python
grouped_data = data.groupby("listed_in(type)")['votes'].sum()
result = pd.DataFrame({"votes": grouped_data})
plt.plot(result.index, result["votes"], c="violet", marker="o")
plt.xlabel("Restaurant Type", color="Green", fontsize=20)
plt.ylabel("Votes", color="Green", fontsize=20)
plt.title("Votes by Restaurant Type", fontsize=18)

plt.show()

```

## **Conclusion-Dinning Resturants Has Received Maximum Votes**

**Pie Chart for book_table**

```python
table_counts = data['book_table'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(table_counts, labels=table_counts.index, autopct='%1.1f%%', startangle=90, colors=['red','orange'])
plt.title("Distribution of Table Booking")
plt.show()
```

# ** Conclusion: The pie chart shows the proportion of restaurants that offer table booking compared to those that do not.**

**Histogram Plot**

```python
plt.hist(data["rate"], bins=5, color="Indigo", edgecolor="black")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Rating Distribution")
plt.show()
```

## **Conclusion-Majority Resturant Received Rating From 3.75 to 4**

**Box Plot**

```python
plt.figure(figsize=(5,6))
sns.boxplot(x="online_order", y="rate", data=data, palette=["blue", "green"])
plt.title("Boxplot of Ratings by Online Order")
plt.show()
```

# **Conclusion: Restaurants offering online orders generally show a different distribution of ratings compared to those that don’t.**

**Heat Map**

```python
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].astype(str).str.replace(',','')
data['approx_cost(for two people)'] = pd.to_numeric(data['approx_cost(for two people)'], errors='coerce')
numeric_data = data[['online_order','book_table','rate','votes','approx_cost(for two people)']]
corr = numeric_data.corr()
plt.figure(figsize=(5,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap of Zomato Dataset", fontsize=16)
plt.show()
```

# The heatmap shows the correlations among online orders, table booking, ratings, votes, and cost.

**Pie Chart for listed_in(type)**

```python
type_counts = data['listed_in(type)'].value_counts().head(5)

plt.figure(figsize=(5,5))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Restaurant Types")
plt.show()
```

#**Conclusion: The pie chart highlights the top 5 restaurant types, showing which categories are most popular in the dataset.**

**Pie Chart for online_order**

```python

online_counts = data['online_order'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(online_counts, labels=online_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue','lightgreen'])
plt.title("Distribution of Online Orders")
plt.show()
```

# Conclusion: The pie chart illustrates the percentage of restaurants that provide online ordering compared to those that do not.
